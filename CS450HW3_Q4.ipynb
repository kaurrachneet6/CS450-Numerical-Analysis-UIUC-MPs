```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse
```

```python
size = 20
cond_no=[]
cond=[]
```

```python
for i in range(size):
    A= np.triu(np.eye(i+1))
    A[np.triu_indices(A.shape[0], 1)] = -1
    M=np.dot(A,A.T)
    #print (np.dot(A,A.T),np.dot(A.T,A))
    eig,vec=np.linalg.eig(M)
    cond.append(np.sqrt(eig.max()/eig.min()))
    #print (vec)
    U, s, V = np.linalg.svd(A)
    #print(U)
    cond_no.append(s.max()/s.min())
```

```python
x=list(range(1,size+1))
plt.plot(x,cond_no,'b')
plt.plot(x,cond,'r')
plt.xticks(x,x)
plt.xlabel('Size of the matrix')
plt.ylabel('Condition no.')
plt.title('Change on condition no. with size of the matrix for nearly singular matrix')
plt.show()
```
