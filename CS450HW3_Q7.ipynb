```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
true_root=2.  #True root of the function
x0=5  #Initial point
iterations =10  #No. of iterations to perform
err=[[] for _ in range(4)] #Relative errors for the four iterative schemes
```

```python
def g(x,i):
    if(i==1):
        return (x**2.+2.)/3.       #g1(x)
    if(i==2):
        return np.sqrt(3.*x-2.)   #g2(x)
    if(i==3):
        return 3.-(2./x)           #g3(x)
    if(i==4):
        return (x**2.-2.)/(2.*x-3.)  #g4(x)
    
```

```python
for i in range(1,5):
    for k in range(iterations):
        x1=g(x0,i)
        (err[i-1]).append(np.abs((x1-x0)/x0))
        x0=x1
```

```python
x=list(range(iterations))
plt.semilogy(x,err[0],'r',label='g1(x)=(x**2.+2.)/3. ')
plt.semilogy(x,err[1],'b',label='g2(x)=np.sqrt(3.*x-2.)')
plt.semilogy(x,err[2],'g',label='g3(x)=3.-(2./x)')
plt.semilogy(x,err[3],'m',label='g4(x)= (x**2.-2.)/(2.*x-3.)')
plt.xlabel('Iterations')
plt.ylabel('Relative error')
plt.title('Fixed point iteration schemes')
plt.legend()
plt.show()
```

```python
print ('It is evident from the above plot that g1(x) is divergent fixed point iterative scheme\nHence, we look at g2(x), g3(x) and g4(x) in the next plot')
```

```python
x=list(range(iterations))
plt.semilogy(x,err[1],'b',label='g2(x)=np.sqrt(3.*x-2.)')
plt.semilogy(x,err[2],'g',label='g3(x)=3.-(2./x)')
plt.semilogy(x,err[3],'m',label='g4(x)= (x**2.-2.)/(2.*x-3.)')
plt.xlabel('Iterations')
plt.ylabel('Relative error')
plt.title('Fixed point iteration schemes')
plt.legend()
plt.show()
```

```python
print ('It is evident from the above plot that g2(x) and g3(x) are linearly convergent fixed point iterative schemes\nAnd g4(x) is quadratically convergent fixed point iterative scheme')
```
